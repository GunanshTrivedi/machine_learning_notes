Naive Bayes Classifier --

Naive bayes theorm uses bayes theorm for conditional probability with a naive assumption that the features are not correlated to each other and tries to find conditional probability of target variable given the probabilities of features. We will use titanic survival dataset here and using naive bayes classifier find out the survival probability of titanic travellers.

There are three types of Naive Bayes Classifier --

1. Bernoulli Naive Bayes :
When we have the features columns in binary such that 0 and 1 .

2. Multinomial Naive Bayes :
It is used when we have discrete data .

3. Gaussian Naive Bayes :
Gaussian Naive Bayes is used in cases when all our features are continuous. 

```python
import pandas as pd 
```

```python
Data = pd.read_csv('titanic.csv')
Data.head(10)
```

Deleting some columns which are not useful --

```python
Data.drop(['PassengerId','Name','SibSp','Parch','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
print(Data)
```

```python
target = Data.Survived
input = Data.drop('Survived',axis='columns',inplace=True)
```

```python
input=Data
```

```python
print(input)
```

```python
print(target)
```

Now converting the string value into numeric value because training only accept numeric value -

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Data.Sex = le.fit_transform(Data.Sex)
Data
```

```python
Data.drop('Ticket',axis='columns',inplace=True)
```

```python
Data
```

```python
input
print(input)
```

```python
input.Age = input.Age.fillna(input.Age.mean())
print(input)
```

Now Deviding the data in parts to train --

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input,target,test_size=0.2)
```

Now Creating the Naive Bayes Classifier --

It will automatically check the probability and train the model according to it.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[1,0,20,50]])
```

Now importing another Dataset ---

To predict the spam emails

```python
df = pd.read_csv('spam.csv')
df
```

```python
df.Category = le.fit_transform(df.Category)
df
```

Now Deviding the data in parts --

```python
x_train,x_test,y_train,y_test = train_test_split(df.Message,df.Category,test_size=0.2)
```

Now Using Count Vectorization to convert the Message column in  numerical value because model cannot take string value--

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(x_train.values)

```

We are using Multinomial Naive Bayes here because the data which is to be trained is in discrete data format so here multinomial naive bayes is used.

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

Now predicting the spam emails --

```python
emails = [
  'Hey mohan , can we get together to watch football game tommorow?',
  'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!',
  
]
emails_count = v.transform(emails)
model.predict(emails_count)
## 1 mail is not spam and 2 mail is spam.
```

Now we need to convert the x_test also in numerical value using the CountVectorizer because it takes only numerical value--

```python
x_test_count = v.transform(x_test)
model.score(x_test_count,y_test)
```

Now using pipeline to easy this process--

```python
from sklearn.pipeline import Pipeline

pip = Pipeline([
  ('vectoizer', CountVectorizer()),
  ('nb',MultinomialNB())
])
```

Now by using Pipeline we have converted the data in numerical value and the data is prepared so no need to do all the process now only make a pipeline always --

```python
pip.fit(x_train,y_train)
```

```python
pip.score(x_test,y_test)
```

```python
pip.predict(emails)
## We got the same results with this process because Pipeline converts the message in numerical value automatically.
```
