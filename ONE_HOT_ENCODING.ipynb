USE TO DEAL WITH DATA WHICH HAS NO RELATION BETWEEN THEM ALSO KNOWN AS NOMINAL DATA


```python
import pandas as pd 
```

```python
Data = pd.read_csv('carprices.csv')
print(Data)
```

MAKING DUMMY DATA USING PANDAS--

```python
dummies = pd.get_dummies(Data.Car)
print(dummies)
```

```python
merged = pd.concat([Data,dummies],axis='columns')
print(merged)
```

```python
Data_final=merged.replace({True:1,False:0})
print(Data_final)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

NOW MAKING X AND Y -- TO USE IN MODEL--

```python
X = Data_final.drop('Price',axis='columns')
print(X)
```

```python
final = X.drop('Car',axis='columns')
print(final)
```

```python
x_data=final
```

```python
y_data = Data_final.Price
print(y_data)
```

Now training Data--

```python
model.fit(x_data,y_data)
```

USING THE MODEL--

According to data we have to put values in prediction as first value will be for mileage and second for Age and third for if audi than 1 or else 0 , fourth if BMW THAN 1 or else 0 , fifth if mercedes then 1 or else 0.

```python
model.predict([[59000,2,1,0,0]])
```
