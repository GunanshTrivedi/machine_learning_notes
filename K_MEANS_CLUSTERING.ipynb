K-MEANS-CLUSTERING -


(For deviding the data into clusters.)

There are three types of Machine learning algorithm --
1. Supervised  -- Where we have our target variable
2. Unsupervised --Where we do not have our target variable
3. Reinforcement 

K means clustering is a type of unsupervised machine learning algorithm where  it make clusters of data.

here k is a centre point define by us which means if there are two clusters than we have to take k =2 and it means there will be two random points which will be near to clusters. 1 point will be near to 1 cluster and second point will be near to 2 cluster.

Method ---  We take random points which is K here than we find the center point of clusters and then put that random point in the center of the cluster of Data.

Now Understanding with an example--

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('income.csv')
Data
```

```python
plt.scatter(Data['Age'],Data['Income($)'])
```

We can see that there are 3 clusters of data so we will take K = 3.

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
predicted=km.fit_predict(Data[['Age','Income($)']])
predicted
```

```python
Data['cluster'] = predicted
Data
```

The Data is devided into 3 parts and it is trained according to this and devided in categories 0 , 1 , 2 .
