WHEN WE HAVE A DATA THEN WE SHOULD TRAIN THE MODEL BY USING 80% OF THE AVAILABLE DATA AND THEN TEST THE MODEL USING 20% DATA.
THIS IS A GOOD PRACTICE.

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('Car_detail.csv')
print(Data)
```

```python
plt.scatter(Data['Mileage'],Data['Price'])
```

```python
plt.scatter(Data['Age'],Data['Price'])
```

WE CAN CLEARLY SEE A LINEAR RELATION BETWEEN THE PRICE AND OTHER PARAMETERS SO WE WILL USE LINEAR REGRESSION HERE.

```python
X = Data[['Mileage','Age']]
Y = Data[['Price']]
```

NOW IMPORTING TRAIN AND TEST SPLIT METHOD --

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size = 0.2)
## here test_size means percent of data used to test.
```

THIS WILL SPLIT THE DATA SET INTO 4 PARTS -
X_ TRAIN = DATA TRAINED FOR X,
X_TEST = DATA which is to be TESTED,
Y_TRAIN = DATA TRAINED FOR Y,
Y_TEST = DATA which is to be tested

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn import linear_model
```

NOW USING THE DATA WHICH IS FOR TRAINING THE MODEL X_TRAIN AND Y_TRAIN

```python
model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
```

NOW USING THE DATA WHICH IS FOR TESTING--

```python
g=model.predict(X_test)
print(X_test)
print(g)
```

NOW THE DATA IS TRAINED FOR X_TRAIN AND Y_TRAIN SO IT WILL GIVE HIGH SCORE FOR THESE BECAUSE THE MODEL IS TRAINED FOR IT.. BUT FOR X_TEST AND Y_TEST THE MODEL IS NOT TRAINED SO IT WILL GIVE SOME LOW ACCURACY, BECAUSE THESE ARE NEW FOR THE MODEL.

```python
model.score(X_test,Y_test)
```
