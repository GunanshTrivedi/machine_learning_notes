BAGGING (ENSEMBLE LEARNING) --

For  converting under fit and overfit data into balanced fit we use this method.

Process --

We take a sample of data from the original data and make a subset of sample data and then take random value in that subset.

We create n number of multiple subsets containing random data .

example --
data = [1,2,3,4,5,6,7,4,5,6,1,2,3]
subset1 = [1,5,7]
subset2 = [6,2,6]
subset3 = [3,4,2]

And then on individual subset we run our machine learning model.

Then we get a result from all the individual model and then we take the majority result or average.

Example -- subset1_score + subset2_score + subset3_score

It is made of two words Bootstrapping and Aggregation.

When we use Bagging then we get Low Bias And Low variance model because diffferent data is sent to every model and it get trained on diffferent datasets so if their is a change of data then it does not effects the model.

- Use Bagging every time whenever make a project and specially in LOW BIAS HIGH VARIANCE Data.

```python
import pandas as pd 
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.isna().sum()
```

```python
x = df.drop('HeartDisease',axis = 'columns')
y = df.HeartDisease
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.Sex = le.fit_transform(df.Sex)
df.ChestPainType = le.fit_transform(df.ChestPainType)
df.RestingECG = le.fit_transform(df.RestingECG)
df.ExerciseAngina = le.fit_transform(df.ExerciseAngina)
df.ST_Slope = le.fit_transform(df.ST_Slope)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores=cross_val_score(DecisionTreeClassifier(),x,y,cv = 5) ## cv = No. of folds
scores.mean()
```

Now using Bagging Classifier --

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
bag_model=BaggingClassifier(
  estimator = DecisionTreeClassifier(),
  n_estimators=100, ## No. of subsets
  max_samples=0.8,
  oob_score=True,
  random_state=0
)
bag_model.fit(x_train,y_train)
bag_model.oob_score_
```

```python
bag_model.score(x_test,y_test)
```

```python
bag_model = BaggingClassifier(
  estimator=DecisionTreeClassifier(),
  n_estimators=100,
  max_samples=0.8,
  oob_score=True,
  random_state=0

)
scores = cross_val_score(bag_model,x,y,cv = 5)
scores.mean()
```

We can see that using normal Decision Tree Classifier we got 0.73 score and by using bag_model we got 0.80 score.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
scores = cross_val_score(SVC(),x,y,cv = 5)
scores.mean()
```

```python
bag_model = BaggingClassifier(
  estimator=SVC(),
  n_estimators=100,
  max_samples=0.8,
  oob_score=True,
  random_state=0
)
bag_model.fit(x_train,y_train)
```

```python
bag_model.score(x_test,y_test)
```

```python
scores = cross_val_score(bag_model,x,y,cv=5)
scores.mean()
```
